# Divide un file CSV in due file CSV

```python
import pandas as pd
```

```python
# Training Set Fraction
TS_FRAC = 0.8
```

```python
# Load into a DataFrame
df = pd.read_csv('appointment_data.csv')

df # Show the data
```

```python
# Shuffle data
df = df.sample(frac=1, random_state=42).reset_index(drop=True) # Don't panic

df # Show after shuffling
```

```python
# Total rows
nrows = df.shape[0]
```

```python
# Training set size
trows = round(nrows * TS_FRAC)
```

```python
# Save Training Set
df.iloc[:trows].to_csv('train.csv', index=False)
```

```python
# Save Test Set
df.iloc[trows:].to_csv('test.csv', index=False)
```
